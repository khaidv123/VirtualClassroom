```python
# Practical Math
PROBLEM = """Một doanh nghiệp tư nhân A chuyên kinh doanh xe gắn máy các loại. Hiện nay,
doanh nghiệp đang tập trung chiến lược vào kinh doanh xe honda Future Fi với chi
phí mua vào là 27 (triệu đồng) và bán với giá 31 (triệu đồng) mỗi chiếc. Với giá bán
này thì số lượng xe mà khách hàng sẽ mua trong một năm là là 600 chiếc. Nhằm
mục tiêu đẩy mạnh hơn nữa lượng tiêu thụ dòng xe đang ăn khách này, doanh
nghiệp dự định giảm giá bán và ước tính rằng nếu giảm 1 (triệu đồng) mỗi chiếc thì
số lượng xe bán ra trong một năm sẽ tăng thêm 200 chiếc. Vậy doanh nghiệp phải
định bán với giá bán mới là bao nhiêu triệu đồng để sau khi đã thực hiện giảm giá,
lợi nhuận thu được sẽ là cao nhất?"""
```

```python
SOLUTION = """
**Bước 1: Gọi ẩn và lập biểu thức lợi nhuận**

Gọi $x$ là số triệu đồng giảm giá mỗi chiếc xe (điều kiện $x > 0$).

* Giá bán mới mỗi xe:

  $$
  31 - x \quad \text{(triệu đồng)}
  $$

* Giá mua vào:

  $$
  27 \quad \text{(triệu đồng)}
  $$

→ **Lợi nhuận trên mỗi xe**:

$$
(31 - x) - 27 = 4 - x \quad \text{(triệu đồng)}
$$

* **Số xe bán được** mỗi năm sẽ là:

$$
600 + 200x \quad \text{(chiếc)}
$$

→ **Tổng lợi nhuận** thu được là:

$$
S(x) = (4 - x)(600 + 200x) \quad \text{(triệu đồng)}
$$

---

#### **Bước 2: Khai triển và rút gọn biểu thức $S(x)$**

$$
S(x) = (4 - x)(600 + 200x)
= 2400 + 800x - 600x - 200x^2
= 2400 + 200x - 200x^2
$$

Vậy:

$$
S(x) = -200x^2 + 200x + 2400
$$

---

#### **Bước 3: Dùng đạo hàm để tìm cực đại của hàm lợi nhuận**

Ta lấy đạo hàm của $S(x)$ theo $x$:

$$
S'(x) = \frac{d}{dx}(-200x^2 + 200x + 2400) = -400x + 200
$$

Tìm điểm cực đại bằng cách giải phương trình $S'(x) = 0$:

$$
-400x + 200 = 0 \Rightarrow x = 0.5
$$

Vì hệ số $x^2$ là âm nên hàm số đạt cực đại tại $x = 0.5$.

---

#### **Bước 4: Tính giá bán tương ứng**

Giá bán mới:

$$
31 - x = 31 - 0.5 = 30.5 \quad \text{(triệu đồng)}
$$

---

### **Bước 5: Kết luận:**

Doanh nghiệp nên **giảm giá 0.5 triệu đồng mỗi xe**, tức là bán với giá **30.5 triệu đồng/chiếc**, để đạt được **lợi nhuận tối đa**.

"""
```

```python
SOLUTION = """
**Bước 1: Xác định số lượng học sinh nhận mỗi loại phần thưởng**

Gọi $x$ là số học sinh nhận phần thưởng là sách Toán và Lý. Khi đó:

*   Số học sinh nhận phần thưởng Toán và Hóa là $7 - x$.
*   Số học sinh nhận phần thưởng Lý và Hóa là $6 - x$.

Tổng số sách Hóa được phát là 5, do đó:
$$5 = (7 - x) + (6 - x) \Rightarrow 5 = 13 - 2x \Rightarrow 2x = 8 \Rightarrow x = 4$$

Vậy:

*   Số học sinh nhận Toán và Lý: $x = 4$
*   Số học sinh nhận Toán và Hóa: $7 - x = 7 - 4 = 3$
*   Số học sinh nhận Lý và Hóa: $6 - x = 6 - 4 = 2$

**Bước 2: Tính kích thước không gian mẫu $n(\Omega)$**

Không gian mẫu là số cách chia phần thưởng cho 9 học sinh sao cho có 4 học sinh nhận Toán và Lý, 3 học sinh nhận Toán và Hóa, 2 học sinh nhận Lý và Hóa.

$$n(\Omega) = C_9^4 \cdot C_5^3 \cdot C_2^2 = \frac{9!}{4!5!} \cdot \frac{5!}{3!2!} \cdot \frac{2!}{2!0!} = 126 \cdot 10 \cdot 1 = 1260$$

**Bước 3: Tính số trường hợp thuận lợi cho biến cố**

Gọi $A$ là biến cố "hai học sinh A, B nhận được phần thưởng giống nhau". Ta xét các trường hợp sau:

*   **Trường hợp 1: A và B cùng nhận Toán và Lý:**

    *   Chọn 2 quyển Toán từ 7 quyển: $C_7^2 = \frac{7!}{2!5!} = 21$
    *   Chọn 2 quyển Lý từ 6 quyển: $C_6^2 = \frac{6!}{2!4!} = 15$
    *   Sau khi A và B đã nhận Toán và Lý, còn lại 2 học sinh khác cũng nhận Toán và Lý.  Số cách chọn 2 học sinh từ 7 học sinh còn lại là $C_7^2=21$.
    *   Số cách phát số phần thưởng còn lại là $C_5^3 C_2^2 = 10$

    Số cách chọn trong trường hợp này là : $C_7^2 . C_6^2 . C_7^2 . C_5^3 .C_2^2 = 21 . 15 . 21 . 10 . 1 = 66150$

    Cách này đang bị sai vì  các cuốn sách Toán/Lý là giống nhau nên không quan tâm đến việc chọn cụ thể 2 quyển Toán hoặc 2 quyển Lý. Ta chỉ cần quan tâm đến việc chọn 2 vị trí học sinh A và B mà thôi.

    Chọn 2 học sinh A và B từ 9 học sinh thì chỉ có 1 cách thôi (vì đã xác định là A và B rồi). Vậy, cần chọn 2 học sinh nữa từ 7 học sinh còn lại để nhận Toán và Lý: $C_7^2 = 21$.
    Sau đó, chọn 3 học sinh từ 5 học sinh còn lại để nhận Toán và Hóa: $C_5^3 = 10$.
    Cuối cùng, 2 học sinh còn lại nhận Lý và Hóa: $C_2^2 = 1$.
    Số trường hợp thuận lợi là: $C_7^2 \cdot C_5^3 \cdot C_2^2 = 21 \cdot 10 \cdot 1 = 210$

*   **Trường hợp 2: A và B cùng nhận Toán và Hóa:**

    *   Chọn 2 học sinh từ 7 học sinh còn lại để nhận Toán và Lý: $C_7^4= 35$.
    *   Còn lại 1 học sinh nhận Toán và Hóa.  Vậy chọn học sinh từ 5 học sinh còn lại: $C_5^1=5$.
    *   Chọn 2 học sinh nhận Lý và Hóa: $C_4^2=6$.

    Chọn 2 học sinh A và B thì chỉ có 1 cách thôi.
    Chọn 1 học sinh từ 7 học sinh còn lại nhận Toán và Hóa: $C_7^1=7$.

    Chọn 4 học sinh từ 6 học sinh còn lại nhận Toán và Lý: $C_6^4 = 15$.
    Còn lại 2 học sinh nhận Lý và Hóa: $C_2^2=1$.

    Số trường hợp thuận lợi là: $C_7^1 \cdot C_6^4 \cdot C_2^2 = 7 \cdot 15 \cdot 1 = 105$

*   **Trường hợp 3: A và B cùng nhận Lý và Hóa:**

    *   Chọn 2 học sinh từ 7 học sinh còn lại để nhận Toán và Lý: $C_7^4=35$.
    *   Còn lại 3 học sinh nhận Toán và Hóa: $C_3^3=1$.

    Chọn 4 học sinh từ 7 học sinh còn lại nhận Toán và Lý: $C_7^4 = 35$.
    Chọn 3 học sinh nhận Toán và Hóa. Số cách chọn là $C_3^3 = 1$.
    Số trường hợp thuận lợi là: $C_7^4 \cdot C_3^3 = 35 \cdot 1 = 35$

Tổng số trường hợp thuận lợi là: $n(A) = 210 + 105 + 35 = 350$

**Bước 4: Tính xác suất**

Xác suất để A và B nhận được phần thưởng giống nhau là:

$$P(A) = \frac{n(A)}{n(\Omega)} = \frac{350}{1260} = \frac{35}{126} = \frac{5}{18}$$

**Kết luận:**

Xác suất để hai học sinh A, B nhận được phần thưởng giống nhau là $\frac{5}{18}$."""
```
